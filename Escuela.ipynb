```python
import pandas as pd
import pandas_ta as ta
from sklearn.ensemble import RandomForestClassifier
import numpy as np
csv = "data/apr2_1h_30k.csv"
info = pd.read_csv(csv, parse_dates = ["Date"], index_col = "Date").dropna()
```

```python
# Paso 1: Calculamos las estrategias y nuestra posición objetivo
# Nota: que varios indicadores tienen espacios vacíos
info.ta.strategy()
```

```python
info['Target'] = (info['Close'] >= info['Open']).astype(int)
info
```

```python
# Paso 2: Calculamos la matriz de correlación
# Nota: Los valores vacíos no se toman en cuenta en el cálculo de la matriz
# Note que algunos valores están vacíos, esto no supone un problema
correlation_matrix = info.corr()
correlation_matrix
```

```python
# Paso 3: Eliminamos indicadores altamente correlacionados
threshold = 0.7  # umbral de correlación
correlated_features = set()
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            colname = correlation_matrix.columns[i]
            correlated_features.add(colname)
            break

relevant_indicators = info.drop(correlated_features, axis=1, inplace=False)
relevant_indicators
```

```python
relevant_indicators2 = relevant_indicators.iloc[100:].dropna(axis=1, how="any")
relevant_indicators2
```

```python
df = pd.DataFrame({
    'A': [1, 0, np.nan, np.nan],
    'B': [5, np.nan, np.nan, 8],
    'C': [9, 10, 11, 12],
    'D': [13, 14, 10, 16]
})
df.corr()
```

```python
from sklearn.feature_selection import f_classif

X = relevant_indicators2
y = info["Target"].iloc[100:]
# Calcula la puntuación F y los p-values para cada característica
f_scores, p_values = f_classif(X, y)

# Crea un DataFrame con las puntuaciones F y los p-values
feature_scores = pd.DataFrame({'Feature': X.columns, 'F_Score': f_scores, 'P_Value': p_values})

# Selecciona características con p-values significativos (por ejemplo, p < 0.05)
significant_features = feature_scores[feature_scores['P_Value'] < 0.05]

# Puedes usar 'significant_features' para seleccionar las características importantes
```
